```python
import http.client
import json
import pandas as pd
import time  # Optional: to avoid rate limit issues

# --- Config ---
asins = ["B07ZPKBL9V", "B07ZPKG5TG", "B084HMH875", "B081Y45G46", "B084H7Q8BQ", "B081Y5D817", "B07ZPJWGKZ", "B084HLYBM6"]
api_host = "real-time-amazon-data.p.rapidapi.com"
headers = {
    'x-rapidapi-key': "cb52439a72msh5cb2aba653aeca8p14bed0jsn3316c17d15ef",
    'x-rapidapi-host': api_host
}

# --- Data Collection ---
all_data = []

for asin in asins:
    try:
        print(f"Fetching data for ASIN: {asin}")
        conn = http.client.HTTPSConnection(api_host)
        conn.request("GET", f"/product-details?asin={asin}&country=US", headers=headers)
        res = conn.getresponse()
        data = res.read()
        conn.close()

        json_data = json.loads(data.decode("utf-8"))
        product_data = json_data.get("data", {})

        if product_data:  # Only append if data is not empty
            df_flat = pd.json_normalize(product_data, sep='_')
            all_data.append(df_flat)

        time.sleep(1)  # Pause to avoid hitting rate limits

    except Exception as e:
        print(f"Error fetching ASIN {asin}: {e}")

# --- Combine All into One DataFrame ---
if all_data:
    df_all = pd.concat(all_data, ignore_index=True)
    pd.set_option('display.max_columns', None)
    print(df_all.head())

    # Save to CSV
    df_all.to_csv("all_amazon_products.csv", index=False)
else:
    print("No data was fetched.")
```
